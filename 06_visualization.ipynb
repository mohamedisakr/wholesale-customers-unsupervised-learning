```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
wholesale_df = pd.read_csv("raw_data\wholesale-customers.csv")
```

### Univariate Plots

###### 3 techniques to understand each feature independently.
- Histograms.
- Density Plots.
- Box and Whisker Plots.

```python
# histogram
wholesale_df.hist(figsize=(20,20))
plt.show()
```

```python
# density plots
wholesale_df.plot(kind='density', subplots=True, layout=(3,3), sharex=False)
plt.show()
```

```python
# box whisker plots (boxplots)
wholesale_df.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False)
plt.show()
```

### Multivariate Plots

###### 2 common plots that show the interactions between multiple variables are 
- Correlation Matrix Plot.
- Scatter Plot Matrix.

```python
# correlation matrix plot
plt.figure(figsize=(12, 12))
sns.heatmap(wholesale_df.corr(), annot=True, cmap='viridis')
```

```python
# scatter plot matrix
pd.plotting.scatter_matrix(wholesale_df)
plt.show()
```
